```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Month":["Jan","Feb","Mar","apr","May"],
    "Sales":[20000,25000,18000,30000,28000],
    "Profit":[5000,7000,4000,9000,8000]
}
df=pd.DataFrame(data)
df
```

```python
print("total sales:",df["Sales"].sum())
print("Average Profit:",df["Profit"].mean())
```

```python
plt.plot(df["Month"],df["Sales"])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
df["Expenses"]=[10000,15000,20000,17000,14000]
df
```

```python
df["Net profit"]=df["Sales"]-df["Expenses"]
df
```

```python
plt.plot(df["Month"],df['Net profit'])
plt.title('Monthly profit')
plt.xlabel("Month")
plt.ylabel("Net profit")
plt.show()
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
df.describe()
```

```python
df.groupby("day")["total_bill"].sum()
```

```python
df.columns
```

```python
print('total revenue:',df["total_bill"].sum())
```

```python
print('average tip:',df["tip"].mean())
```

```python
df.groupby('smoker')['tip'].sum()
```

```python
df.groupby('day')['total_bill'].sum().plot(kind='bar')
plt.title("revenue by day")
plt.show()
```

```python
df.groupby('sex')["tip"].mean().plot(kind='bar')
plt.title('average tip by gender')
plt.xlabel('sex')
plt.ylabel('tip')
plt.show()
```

```python
print("Insights")
print("weekend days generate higher revenue compared to other days'")
print("smokers likely to give slightly moe tips than non smokers")
```

```python
df['tip_percent']=(df["tip"]/df['total_bill'])*100
df.head(100)
```

```python
df.groupby("day")['total_bill'].sum()
```

```python
sales_by_day=df.groupby("day")['total_bill'].sum()
print("highest revenue by day:",sales_by_day.idxmax())
print("lowest revenue by day:",sales_by_day.idxmin())
```

```python
df.groupby('day')['tip_percent'].mean()
```

```python
df.groupby('day')['tip_percent'].mean().plot(kind='bar')
plt.title("Average Tip Percentage By Day")
plt.show()
```

```python
print('Insights;')
print('1: the highest revenue day is on saturday.')
print('2:the lowest revenue day is on friday which is one day before the highest revenue day.')
print('3:the tip percentage is quit unpredictable,it varies accross the days')
```
